Does a Higher BMI and being a Smoker increase insurance cost?


Insurance cost tends to rise based of the individuals health status; usually if they are more of a risk to the insurance company, then the patient would have an increased insurance cost.
In this project, I'm going to see if the insurance cost does increase based off of the patients bmi and smoking status.


```python
age = []
sex = []
bmi = []
children = []
smoker = []
region = []
insurance_cost = []
```

```python
with open('insurance.csv') as insurance_csv:
    insurance_reader = csv.DictReader(insurance_csv)
    for row in insurance_reader:
        age.append(row['age'])
        sex.append(row['sex'])
        bmi.append(float(row['bmi']))
        children.append(row['children'])
        smoker.append(row['smoker'])
        region.append(row['region'])
        insurance_cost.append(float(row['charges']))
```

```python
def create_dictionary(age, sex, bmi, children, smoker, region, insurance_cost):
    patient_info ={}
    for element in range(len(age)):
        patient_info[element] = { 'age': age[element],
                                  'sex': sex[element],
                                  'bmi': bmi[element], 
                                  'children': children[element], 
                                  'smoker': smoker[element], 
                                  'region': region[element], 
                                  'insurance cost': insurance_cost[element] 
        }
    return patient_info
patient_records = create_dictionary(age, sex, bmi, children, smoker, region, insurance_cost)
print(patient_records)
```

The function below shows that being a smoker will have a higher insurance cost than a non-smoker.


```python
smoking_cost = list(zip(smoker,insurance_cost))
def smoker_nonsmoker(smoking_cost):
    smoker_count = 0
    nonsmoker_count = 0
    smoker_cost = 0
    nonsmoker_cost = 0
    for element in smoking_cost:
        if 'yes' in element:
            smoker_count += 1
            smoker_cost += element[1]
        if 'no'in element:
                nonsmoker_count += 1 
                nonsmoker_cost += element[1] 
    avg_total_smoker =  smoker_cost / smoker_count
    avg_total_nonsmoker = nonsmoker_cost / nonsmoker_count
    diff_in_smoking = abs(avg_total_smoker - avg_total_nonsmoker)
    return round(diff_in_smoking, 2)
print("The difference in the average insurance cost between smoker and non-smoker is: " + str(smoker_nonsmoker(smoking_cost)) + " dollars.")         
```

This function is to show the insurance cost based off of BMI levels. The ranges of BMi is: 18.5 < 25 (healthy weight), 25 < 30 (overweight), 30+ (obesity)

```python
bmi_cost = list(zip(bmi,insurance_cost))
def bmi_level(bmi_cost):
    normal_weight = 0
    normal_weight_cost = 0
    overweight = 0
    overweight_cost = 0
    obese = 0 
    obese_cost = 0
    below_normal_cost = 0
    below_normal = 0

    for element in bmi_cost:

        if element[0] <= 18.4:
            below_normal += 1
            below_normal_cost += element[1]
        if element[0] == 18.5 and element[0] < 27:
            normal_weight+= 1
            normal_weight_cost += element[1]
        if element[0] <= 26 and element[0] < 31:
            overweight += 1
            overweight_cost += element[1]
        if element[0] >= 30:
            obese += 1
            obese_cost += element[1]
    avg_below_weight = round(below_normal_cost / below_normal , 2)
    avg_normal_weight  = round(normal_weight_cost / normal_weight , 2)
    avg_overweight = round(overweight_cost / overweight, 2)
    avg_obesity = round(obese_cost / obese, 2)
    return avg_below_weight, avg_normal_weight, avg_overweight, avg_obesity
avg_below_weight, avg_normal_weight, avg_overweight, avg_obesity = bmi_level(bmi_cost)
print("The average insurance cost of patients that have below normal weight is: " + str(avg_below_weight) + " dollars.")
print("The average insurance cost of patients that have normal weight is: " + str(avg_normal_weight) + " dollars.")
print("The average insurance cost of patients that are overweight is: " + str(avg_overweight) + " dollars.")
print("The average insurance cost of patients that are obese weight is: " + str(avg_obesity) + " dollars.")
```

We can tell that having a higher BMI level will increase the amount of insurance cost.
We can infer that the higher a BMI level is and if the patient is a smoker or not, the insurance cost should increase.


```python
bmi_smoking = list(zip(smoker, bmi, insurance_cost))
def bmi_smoking_cost(bmi_smoking): 
    low_smoke = 0
    mid_smoke = 0
    high_smoke = 0
    low_non_smoke = 0
    mid_non_smoke = 0
    high_non_smoke = 0
    low_smoke_patient = 0
    low_nonsmoke_patient = 0
    mid_smoke_patient = 0
    mid_nonsmoke_patient = 0
    high_smoke_patient = 0
    high_nonsmoke_patient = 0
    for element in bmi_smoking:
        if element[0] == 'yes' and element[1] <= 18.5 and element[1] < 26:
            low_smoke_patient += 1
            low_smoke += element[2]
        elif element[0] == 'no' and element[1] == 18.5 and element[1] < 26:
            low_nonsmoke_patient += 1
            low_non_smoke += element[2]
        elif element[0] == 'yes' and element[1] <= 26 and element[1] < 30:
            mid_smoke_patient += 1
            mid_smoke += element[2]
        elif element[0] == 'no' and element[1] <= 26 and element[1] < 30:
            mid_nonsmoke_patient += 1
            mid_non_smoke += element[2]
        elif element[0] == 'yes' and element[1] >= 30:
            high_smoke_patient += 1
            high_smoke += element[2]
        elif element[0] == 'no' and element[1] >= 30:
            high_nonsmoke_patient += 1
            high_non_smoke += element[2]

    return  round(low_smoke / low_smoke_patient, 2), round(low_non_smoke / low_nonsmoke_patient, 2), round(mid_smoke / mid_smoke_patient, 2), round(mid_non_smoke / mid_nonsmoke_patient, 2), round(high_smoke / high_smoke_patient, 2), round(high_non_smoke / high_nonsmoke_patient, 2)
low_smoke, low_non_smoke, mid_smoke, mid_non_smoke, high_smoke, high_non_smoke = bmi_smoking_cost(bmi_smoking)
print('The average insurance cost of a low-range BMI and smoker is: ' + str(low_smoke) + " dollars")
print('The average insurance cost of a low-range BMI and non-smoker is: ' + str(low_non_smoke) + " dollars")
print('The average insurance cost of a mid-range BMI and smoker is: ' + str(mid_smoke) + " dollars")
print('The average insurance cost of a mid-range BMI and non-smoker is: ' + str(mid_non_smoke) + " dollars")
print('The average insurance cost of a high-range BMI and smoker is: ' + str(high_smoke) + " dollars")
print('The average insurance cost of a high-range BMI and non-smoker is: ' + str(high_non_smoke) + " dollars")
```

Based of the function above we are able to conclude that having a higher BMI and being a smoker will increase your insurance cost.
